### Importando as bibliotecas

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Carregando dados de teste.
test = pd.read_csv(r'C:\Users\jpcar\Documents\DL_Projects\santander-customer-transaction-prediction\test.csv')
# Exibindo as primeiras linhas.
test.head()
```

### redução e transformação de features

```python
X_train = train.drop(['ID_code', 'target'], axis = 1)
X_train.head()
```

```python
y_train = train['target']
y_train.head()
```

```python
X_test = test.drop(['ID_code'], axis = 1)
X_test.head()
```

```python
test.isnull().sum().sum()
```

```python
X_train.var()
```

```python
ss = StandardScaler()
X_train_Scaled = ss.fit_transform(X_train)
```

```python
X_test_Scaled = ss.transform(X_test)
```

```python
y_train.value_counts(dropna=False)
```

```python
encoder = LabelEncoder()
encoder.fit(y_train)
encoded_Y = encoder.transform(y_train)
```

### Configuração da rede 

```python
model = Sequential()
model.add(Dense(100, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train_Scaled, encoded_Y, epochs=100, batch_size = 35, shuffle=True, verbose=1)
```

```python
pred = model.predict(X_test_Scaled)
```

### Predição na base de testes

```python
result = pd.DataFrame({"ID_code": test['ID_code'], "target": pred[:,0]})
result.head()
```


### Arquivo para submissão

```python
result.to_csv("submission1.csv", index=False)
```
